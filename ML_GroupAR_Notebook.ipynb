# 1.Introduction


**1.1 Background and problem statement**.  
Employment attrition is one of the most pressing challenges organizations face. High turnover directly translates into increased
recruitment expenses, loss of institutional knowledge and reduced productivity. For human resource professionals, understanding the
drivers of attrition is essential for developing effective retention strategies and ensuring long term organizational sustainability.

Though numerous elements, such as stress and work-life balance, influence an employee's turnover this research focuses specifically
on two common variables: overtime work and monthly income. While prior studies have extensively examined predictors such as job
satisfaction, stress and work life balance, the interaction between workload and compensation remains underexplored. Although
overtime has been widely identified as a stressor linked to higher attrition, the extent to which income moderates this effect has not
been sufficiently tested.

This oversight is significant because a common organizational assumption is that increased pay can fully compensate for the strain of
long working hours. This research challenges that simplistic view. Financial incentives may not completely offset the negative impacts
of overwork on an employee’s wellbeing and commitment. Analysing this complex relationship will not only provide practical guidance
for organizational policy but will also enrich established theoretical models of motivation and turnover, like the Job Demands-Resources
(JD-R) model and Equity Theory, by introducing compensation as a key moderating or contextual factor.
    

**1.2 Research Objectives, Dataset, and Key Variables**  
This investigation leverages the HR Employee Attrition dataset (comprising 1,470 employees record) to systematically examine how the
combined effect of overtime and monthly income influences turnover. The study is structured around three core research questions:  
1. Is there a significant relationship between working overtime and the probability of employee attrition?  
2. Does lower monthly income statistically increase the likelihood of employee attrition?
3. To what extent does monthly income moderate the relationship observed between overtime and attrition?

The key variables central to this study are:  
* ***Attrition***: The dependent binary variable (coded 1= Yes, 0= No) indicating whether an employee has left the company.  
* ***Overtime***: The primary independent binary variable (coded 1= Yes, 0= No) used as proxy for increased workload. In a Preliminary
univariate logistic regression model, the odds ratio for overtime was approximately 3.6, indicating employees working overtime
were 3.6 times more likely to attrite then those who were not, making it the strongest single predictor identified.
* ***Monthly Income***: A numerical variable functioning as both an independent predictor and the potential moderator. Due to its
highly right-skewed distribution (skewness: 1.37), a logarithmic transformation (using Log2 or Log10) was necessary to
approximate normal distribution and satisfying subsequent statistical modelling assumptions.  

**1.3 Research Gaps and contributions**  
Most research on employee attrition investigates factors like job satisfaction, work–life balance, or stress. Overtime has also been
studied, but income as a moderating factor in this relationship has received limited attention. This study provides analytical depth by
asking how the financial context changes the risk posed by overwork.
  
* ***Organizations Relevance:*** Many companies treat higher pay as a blanket solution to overwork. This research empirically tests
whether higher income truly buffers the negative effects of overtime or if this protective effect diminishes over time.
* ***HR policy Implications:*** The Findings can directly inform and refine compensation structures and
workload management policies, particularly in demanding sectors such as consulting, healthcare, or technology.
* ***Theoretical Advancement:*** The studies enrich turnover theory by formally modelling compensation as a contextual element that
interacts with job demands, specifically contributing to the job Demand Resources (JD−R) and Equity Theory frameworks. This
modelling challenges the fundamental business assumption that monetary rewards can universally justify excessive workload, a
finding that could reshape theoretical perspective on employee well-being and retention dynamics.
  
**1.4 Hypothesis and Testing plan**   
This project aims to quantify the interrelationship between working overtime, monthly income and employee attrition.  
* ***Null Hypothesis (H0):*** Monthly income does not moderate the relationship between overtime and attrition.  
* ***Alternative Hypothesis (H1):*** Higher monthly income reduces the positive effect of overtime on attrition.   

The analysis plan is dived into two phases:
1. ***Exploratory data analysis (EDA):*** This phase involves descriptive statistics and generating visualizations to summarize variable
distributions and preliminary relationships.  
2. ***Inferential statistical methods:***
  * Independent Samples t-test:  Used to compare mean attrition rates between the group of employees who work overtime and
  the group who do not.
  * Logistic Regression Modelling: This model is applied to estimate the probability of attrition as a function of overtime and
  monthly income, including an interaction term to formally test the moderating effect of income.



**Hypothesis**

***Null Hypothesis (H0)*** - Monthly income does not moderate the relationship between overtime and attrition.

***Alternative Hypothesis (H1)*** - Monthly income does moderate the relationship between overtime and attrition.

####3.2.2.3.Absolute and Relative Frequency plots

```python
import matplotlib.pyplot as plt
import pandas as pd

# Function to plot absolute and relative frequency for a given variable
def plot_frequency(series, title):
    # Absolute frequency
    abs_freq = series.value_counts()

    # Relative frequency
    rel_freq = series.value_counts(normalize=True) * 100

    # Plotting
    plt.figure(figsize=(14, 6))

    # Absolute frequency
    plt.subplot(1, 2, 1)
    abs_freq.plot(kind='bar', color='skyblue')
    plt.title(f'Absolute Frequency of {title}')
    plt.xlabel(title)
    plt.ylabel('Count')
    plt.grid(axis='y', linestyle='--')

    # Relative frequency
    plt.subplot(1, 2, 2)
    rel_freq.plot(kind='bar', color='salmon')
    plt.title(f'Relative Frequency of {title} (%)')
    plt.xlabel(title)
    plt.ylabel('Percentage')
    plt.grid(axis='y', linestyle='--')

    plt.tight_layout()
    plt.show()


# Plot OverTime
plot_frequency(df['OverTime'], 'OverTime')

# Plot Attrition
plot_frequency(df['Attrition'], 'Attrition')

# For MonthlyIncome, use quantile-based binning
income_quintiles = pd.qcut(df['MonthlyIncome'], q=5)
plot_frequency(income_quintiles, 'Monthly Income (Quantile Binned)')
```

###3.2.4.Correlation between Attrition and other Attributes

***Attrition will be chosen as our main outcome variable for the following reasons:***

- Business relevance: Employee turnover is one of the most expensive HR challenges. It directly affects recruitment costs, training investment, loss of expertise, and overall productivity. Therefore, attrition is not just another variable but a central KPI for managers.

- Statistical fit: Attrition is a binary Yes/No variable, which makes it ideal for the statistical methods we learned (chi-square, t-tests, logistic regression). Unlike other continuous measures, it defines a clear outcome that can be modeled and predicted.

- Data structure: The dataset shows that attrition is imbalanced (≈16% Yes vs. 84% No). This reflects real-world HR data where most employees stay, but the minority who leave create significant impact. This imbalance makes attrition both realistic and analytically interesting.

- Role in dataset: Attrition represents the outcome we want to explain and predict. Other variables (income, overtime, satisfaction, tenure, incentives) act as possible drivers or influencing factors. Focusing on attrition allows us to connect descriptive insights with actionable recommendations.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# STEP 1: one-hot encode multi-category text columns (nominals)
nominal_cols = [c for c in ["BusinessTravel","Department","EducationField","JobRole","MaritalStatus"] if c in df.columns]
df_corr = pd.get_dummies(df, columns=nominal_cols, drop_first=True)

# STEP 2: keep only numeric columns and exlude logs for correlation analysis
num_df = df_corr.select_dtypes(include=[np.number]).copy()
num_df = num_df[[c for c in num_df.columns if "log" not in c.lower()]]

# STEP 3: compute Spearman correlations with Attrition
corr_spear = num_df.corr(method="spearman")["Attrition_Flag"].sort_values(ascending=False)
corr_tbl = (
    corr_spear.to_frame("corr_spearman")
    .assign(abs_corr=lambda d: d["corr_spearman"].abs())
    .sort_values("abs_corr", ascending=False)
    .drop(index="Attrition_Flag", errors="ignore")
)
print("Top correlations with Attrition:")
display(corr_tbl.head(20))

# STEP 4: bar chart of top correlations (visual ranking)
top_k = 15
top_vars = corr_tbl.head(top_k).index
plt.figure(figsize=(8,6))
sns.barplot(x=corr_tbl.loc[top_vars, "corr_spearman"], y=top_vars, orient="h")
plt.axvline(0, linestyle="--", linewidth=1)
plt.title(f"Top {top_k} Spearman correlations with Attrition")
plt.xlabel("Spearman correlation")
plt.ylabel("Variable")
plt.tight_layout()
plt.show()

# STEP 5: heatmap of top correlated variables (see overlaps/redundancy)
heat_vars = ["Attrition_Flag"] + list(top_vars)
plt.figure(figsize=(9,7))
sns.heatmap(num_df[heat_vars].corr(method="spearman"), annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlation Heatmap (Top variables & Attrition)")
plt.tight_layout()
plt.show()
```

OverTime is the strongest positive correlate of attrition, confirming that employees who work overtime are substantially more likely to quit.
This makes it a interesting variable to analyze.

Tenure and experience variables (YearsAtCompany, TotalWorkingYears, Age) have the strongest negative correlation, suggesting that newer and younger employees are more at risk of leaving.

#4.Method 1 - Preliminary Tests of Attrition and Independent Variables

Before conducting regression analysis, we perform preliminary tests to examine whether key independent variables (OverTime and Monthly Income) are significantly associated with Attrition.

**Chi-Square Test (OverTime × Attrition)**

H₀: OverTime and Attrition are independent (no association).

H₁: OverTime and Attrition are not independent (there is an association).

**Two Sample t-Test (Monthly Income × Attrition)**

H₀: The mean Monthly Income is the same for employees who stayed and those who left.

H₁: The mean Monthly Income differs between employees who stayed and those who left.

```python
import pandas as pd
from scipy.stats import chi2_contingency

contingency = pd.crosstab(df["OverTime_Flag"], df["Attrition_Flag"])
print("Contingency Table:\n", contingency, "\n")

# Run Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency)

print(f"Chi-Square statistic = {chi2:.3f}")
print(f"Degrees of freedom   = {dof}")
print(f"P-value              = {p:.4f}")

if p < 0.05:
    print("Result: Reject H0")
else:
    print("Result: Fail to reject H0")
```

```python
from scipy.stats import ttest_ind


income_stayed = df.loc[df["Attrition_Flag"] == 0, "MonthlyIncome_log2"].dropna()
income_left   = df.loc[df["Attrition_Flag"] == 1, "MonthlyIncome_log2"].dropna()

#Run Two-Sample t-test
t_stat, p_val = ttest_ind(income_stayed, income_left, equal_var=True)

print("Two-Sample t-test: Monthly Income (log2) by Attrition")
print(f"T-statistic = {t_stat:.3f}")
print(f"P-value     = {p_val:.4f}")


if p_val < 0.05:
    print("Result: Reject H0 → Monthly Income differs significantly between attrition groups.")
else:
    print("Result: Fail to reject H0 → No significant difference in Monthly Income between groups.")
```

According to the chi-square test, there is a significant correlation between attrition and overtime (χ²(1) = 87.56, p < 0.001), indicating that workers who put in extra hours are far more likely to quit than those who don't. Likewise, the two-sample t-test revealed a significant difference in mean monthly income between attrition groups (t = 7.748, p < 0.001).Due to monthly income being highly skewed, we conducted the test using log-transformed income. The results remained significant, confirming that the relationship between income and attrition is significant while addressing skewness in the distribution. Employees who left the company earned, on average, less than those who stayed. Together, these findings confirm that both overtime and monthly income are strongly related to attrition, providing support for further analysis of their interaction using logistic regression. As our analysis so far showed overtime is the biggest factor that leads to attrition for employees.We will be investigating whether having higher income levels moderate the positive effect of overtime.




#5.Method 2 - Logistic Regression

##5.1.Logistic Regression

H₀: The interaction between Overtime and Monthly Income is not significant.

H₁: The interaction is significant, such that higher income reduces the effect of overtime on attrition.

```python
import numpy as np
import statsmodels.formula.api as smf
import pandas as pd

#Create centered log-income variable
df["MonthlyIncome_log2_c"] = df["MonthlyIncome_log2"] - df["MonthlyIncome_log2"].mean()

# Fit logistic regression with interaction
model= smf.logit("Attrition_Flag ~ OverTime_Flag * MonthlyIncome_log2_c",data=df).fit()

print(model.summary())
```

The variable log2 transformed Monthly income was used to reduce skewness and capture proportional rather than absolute effects. We additionally mean-centered the log-transformed variable before including it in the interaction term with OverTime.Centering helps interpret coefficients around the average income and avoids the unrealistic baseline (log-income = 0), improving model interpretability without affecting overall model fit.The regression results show that overtime has a strong positive effect on attrition, while monthly income has a strong negative effect. Moreover, the interaction term (Overtime:Monthly Income) is negative, suggesting that higher income buffers the effect of overtime on attrition. Although the interaction is only marginally significant (p ≈ 0.055), the direction aligns with our hypothesis: employees with higher income are less affected by the negative impact of overtime on attrition risk.We find partial evidence to reject H₀.

## 5.2.Odds Ratio

```python
import numpy as np
import pandas as pd

# Get coefficients and confidence intervals
params = model.params
conf = model.conf_int()

# Exponentiate to get odds ratios
odds_ratios = np.exp(params)
conf_odds = np.exp(conf)

# Combine into a table
or_table = pd.DataFrame({"Odds Ratio": odds_ratios})

print("Odds Ratios with 95% Confidence Intervals:")
display(or_table)
```

Before centering the log Monthly income varible we found the overtime variable OR close to 300, after centering the log Income variable Odds Ratio became interpretable.At average income, employees who work overtime are about 3.7 times more likely to leave the company. The odds of attrition decrease as income rises (OR = 0.59), and the negative interaction (OR = 0.71) indicates that the effect of overtime is stronger for lower-income employees. In other words, high-income employees seem more tolerant of overtime, while low-income employees are more likely to quit when required to work overtime.



##5.3.Goodness of Fit

 ### 5.3.1.Pseudo R-squared Metrics

```python

# McFadden
mcfadden_r2 = 1 - (model.llf / model.llnull)

# Cox & Snell
n = model.nobs
cox_snell_r2 = 1 - np.exp((model.llnull - model.llf) * 2.0 / n)

# Nagelkerke (correct denominator uses + sign: exp( 2*llnull / n ))
nagelkerke_r2 = cox_snell_r2 / (1 - np.exp(2.0 * model.llnull / n))

print(f"McFadden's pseudo R-squared: {mcfadden_r2:.4f}")
print(f"Cox & Snell pseudo R-squared: {cox_snell_r2:.4f}")
print(f"Nagelkerke pseudo R-squared: {nagelkerke_r2:.4f}")
```

Looking at the different pseudo R² measures, our model explain nearly 10-17% of the variation in attrition. While these values may seem modest compared to linear regression, they are typical in HR and behavioral research, where employee turnover is influenced by many external and unobserved factors. We can say the results confirm that overtime and income are meaningful predictors of attrition, even though they do not capture the full complexity of employee decisions.

### 5.3.2 Classification Table

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Null model
y_pred_null = np.zeros_like(df["Attrition_Flag"])
cm_null = confusion_matrix(df["Attrition_Flag"], y_pred_null)
plt.figure(figsize=(4,3))
sns.heatmap(cm_null, annot=True, fmt="d", cmap="Purples", cbar=False,xticklabels=["Predicted Stay"], yticklabels=["Actual Stay", "Actual Leave"])
plt.title("Null Model Classification Table")
plt.show()

#Full model
y_pred_prob = model.predict(df)
y_pred = (y_pred_prob >= 0.5).astype(int)
cm = confusion_matrix(df["Attrition_Flag"], y_pred)
plt.figure(figsize=(4,3))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,xticklabels=["Predicted Stay", "Predicted Leave"],yticklabels=["Actual Stay", "Actual Leave"])
plt.title("Full Model Classification Table")
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.show()
```

The full model performs better than the null model by correctly identifying some employees who left, which the null model entirely failed to do.
